```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(df.drop('target',axis=1), df['target'], test_size=0.2, random_state=0 ,stratify=df['target'])
```

```python
from lazypredict.Supervised import LazyClassifier
from lazypredict.Supervised import LazyRegressor
```

```python
model = LazyClassifier(verbose=0, ignore_warnings=True,predictions=True )
models, predictions = model.fit(x_train, x_test, y_train,y_test)
models
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions['BernoulliNB']))
```

# Classifier vs Regressor
1. Regressor (회귀)
 - 목표: 연속적인 값(실수)을 예측합니다.
 >예: 주택 가격 예측, 주식 가격 예측, 온도 예측, 소득 수준 예측 등.

 - 출력값: 숫자(실수) 형태의 연속 데이터.
 - 대표적인 알고리즘:
 1. 선형 회귀 (Linear Regression)
 2. 결정 트리 회귀 (Decision Tree Regressor)
 3.랜덤 포레스트 회귀 (Random Forest Regressor)
 4.서포트 벡터 회귀 (SVR )
 5.XGBoost Regressor, CatBoost Regressor 등.

2. Classifier (분류)
 - 목표: **이산적인 클래스(카테고리)**를 예측합니다.
 >예: 이메일 스팸 여부 예측(스팸/비스팸), 암 진단 여부(양성/음성), 이미지 속 물체 분류(고양이/강아지/자동차 등).

 - 출력값: 클래스 레이블(0, 1, 2 등) 또는 확률 값.
 - 대표적인 알고리즘:
 1. 로지스틱 회귀 (Logistic Regression)
 2. 결정 트리 분류 (Decision Tree Classifier)
 3. 랜덤 포레스트 분류 (Random Forest Classifier)
`4. 서포트 벡터 머신 (SVM)
 5. XGBoost Classifier, CatBoost Classifier 등.
