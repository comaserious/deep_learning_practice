```python
from surprise import SVD, Dataset, Reader, accuracy
from surprise.model_selection import train_test_split
from surprise.dataset import DatasetAutoFolds
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/ratings.csv')
df
```

```python
df['rating'].unique()
```

```python
df.to_csv('ratings_nohead.csv', index=False, header=False)
```

```python
reader = Reader(line_format='user item rating timestamp', sep=',', rating_scale=((0.5,5)))
data_folds = DatasetAutoFolds(ratings_file='ratings_nohead.csv', reader = reader)
```

```python
trainset = data_folds.build_full_trainset()

model = SVD(n_factors=50 , random_state = 0, n_epochs =100)
model.fit(trainset)
```

```python
movies = pd.read_csv('/content/sample_data/movies.csv')
movies
```

```python
# @title
def get_unseen_surprise(ratings, movies, userId):
  seen_movies = df[df['userId']==userId]['movieId'].tolist()
  total_movies = movies['movieId'].tolist()

  unseen_movies=[]
  for movie in total_movies:
    if movie not in seen_movies:
      unseen_movies.append(movie)

  return unseen_movies
```

```python
unseen_movies = get_unseen_surprise(df,movies,9)
unseen_movies
```

```python
def recommend_movie_by_surprise(model, userid, unseen_movies , top_n = 5):
    pred_list = []

    for movieid in unseen_movies:
        pred = model.predict(str(userid),str(movieid))
        pred_list.append(pred)

    def sort_key(pred):
        return pred.est

    pred_list.sort(key=sort_key, reverse=True)

    top_pred = pred_list[:top_n]

    result = []
    for temp in top_pred:
        dict_temp= {'영화ID' : temp.iid, 'rating' : temp.est , '영화명' : movies[movies['movieId']==int(temp.iid)]['title']}
        result.append(dict_temp)

    return result
```

```python
recommend_movie_by_surprise(model, 9 , unseen_movies, 5)
```
