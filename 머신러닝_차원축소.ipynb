```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

iris_data = load_iris()
iris_data
```

```python
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris_df['target'] = iris_data.target
iris_df.head()
```

```python
markers = ['^' , 's' , 'o']

for i, marker in enumerate(markers):
  x_data = iris_df[iris_df['target']==i]['sepal length (cm)']
  y_data = iris_df[iris_df['target']==i]['sepal width (cm)']
  plt.scatter(x_data , y_data , marker=marker , label=iris_data.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
for i , marker in enumerate(markers):
  x_data = iris_df[iris_df['target']==i]['petal length (cm)']
  y_data = iris_df[iris_df['target']==i]['petal width (cm)']

  plt.scatter(x_data, y_data ,marker = marker ,label=iris_data.target_names[i])

plt.legend()
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
iris_df
```

```python
scaler = StandardScaler() # normalization
scaler.fit(iris_df.iloc[:,:-1])
iris_scaled = scaler.transform(iris_df.iloc[:,:-1])
iris_scaled
```

```python
pca = PCA(n_components=2) #    2 (  4   2    sepal    )
pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)
iris_pca
```

```python
irispca_df = pd.DataFrame(iris_pca, columns =['pca_component1','pca_component2'])
irispca_df['target'] = iris_data.target
irispca_df.head()
```

```python
irispca_df[irispca_df['target']==1]
```

```python
markers = ['^' , 's', 'o']
for i , marker in enumerate(markers):
  x_data = irispca_df[irispca_df['target']==i]['pca_component1']
  y_data = irispca_df[irispca_df['target']==i]['pca_component2']

  plt.scatter(x_data, y_data , marker = marker, label = iris_data.target_names[i])

plt.legend()
plt.xlabel('sepal length')
plt.ylabel('seaal width')

plt.show()
```
