```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('/content/ratings.csv')
movies = pd.read_csv('/content/movies.csv')

ratings
```

```python
movies
```

```python
movies_ratings = pd.merge(ratings, movies , on ='movieId' )
movies_ratings
```

```python
title_user = movies_ratings.pivot_table(index='userId', columns='title' , values='rating')
title_user = title_user.fillna(0)
title_user
```

# pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All', dropna=True)

 - data: 피벗 테이블을 생성할 원본 데이터프레임.
 - values: 집계할 열(값). 선택 사항이며, 지정하지 않으면 모든 숫자 열이 집계 대상이 됩니다.
 - index: 행 레벨로 사용할 열(축).
 - columns: 열 레벨로 사용할 열(축).
 - aggfunc: 데이터 집계 함수(기본값은 'mean'). 예: 'sum', 'count', 'median', 사용자 정의 함수 등.
 - fill_value: 누락된 값을 대체할 값.
 - margins: 행과 열의 총합(Total)을 포함할지 여부. (기본값은 False)
 - margins_name: margins=True일 때 총합(Total) 행/열의 이름(기본값: 'All').


```python
user_based_result = cosine_similarity(title_user,title_user)
user_based_result
```

```python
user_based_df = pd.DataFrame(user_based_result, index=title_user.index,columns=title_user.index)
user_based_df
```
