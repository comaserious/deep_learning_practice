```python
from sklearn.feature_extraction.text import CountVectorizer
text =[
    "Action Adventure Fantasy ScienceFiction",
    "누구나 한번쯤은 사랑에 울고",
    "그것이 바로 사랑 사랑 사랑이야"
]

count_vec = CountVectorizer()
m = count_vec.fit_transform(text)
count_vec.vocabulary_
print(count_vec.vocabulary_)
m.toarray()
```

CounterVectorizer 는 텍스트를 벡터화 시켜주는 함수이다.
텍스트 빈도수에 기반하여 벡터화시켜준다

fit_transform 으로 단어사전 생성과 벡터화를 진행한다
toarray 는 벡터데이터를 2d 행렬로 변환 시켜줍니다.

```python
count_vec= CountVectorizer(tokenizer=lambda x : x, lowercase =False) # input 값이 list 형식으로 공백기준으로 tokenizer 할필요 없기 때문에 tokenizer = lambda x: x 를 통해서 중복을 피한다
genres_count_result = count_vec.fit_transform(movies_df['genres'])
genres_mat = genres_count_result.toarray()
count_vec.vocabulary_
```

#영화 추천 시스템


```python
df = pd.read_csv('/content/tmdb_5000_movies.csv')
movies_df = df[['id','title','genres','vote_average', 'vote_count','popularity','keywords','overview']]
pd.set_option('display.max_colwidth', 500)
```

```python
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
def extract_names(data):
  return [item['name'] for item in data]

movies_df['genres'] = movies_df['genres'].apply(extract_names)
movies_df['keywords'] = movies_df['keywords'].apply(extract_names)
```

```python
count_vec = CountVectorizer(tokenizer=lambda x : x, lowercase=False)
combined_features = movies_df['genres'] + movies_df['keywords']
count_mat = count_vec.fit_transform(combined_features)
cosine_sim = cosine_similarity(count_mat)
```

```python
cosine_sim
```

```python
movies_df[movies_df['title']=='The Dark Knight'].index
```

```python
def get_recommendations(title, cosine_sim = cosine_sim):

  idx = movies_df[movies_df['title']==title].index[0] # 우선 요청 영화의 인덱스를 검색

  sim_scores = list(enumerate(cosine_sim[idx])) # enumerate 는 (index , value) 로 반환한다

  sim_scores = sorted(sim_scores,key=lambda x: x[1],reverse=True)
  # 정렬하려고 하는 list 의 요소가 튜블일 때 key 매개 변수는 각 튜블의 몇번째 요소를 기준으로 정렬을 할 것인지 정한다 여기서는 두번째 요소를 기준으로 정렬하겠다는 뜻이다

  sim_scores = sim_scores[1:11]

  movies_indicates = [i[0] for i in sim_scores]

  return movies_df['title'].iloc[movies_indicates]

print(get_recommendations('Deadpool'))
```
